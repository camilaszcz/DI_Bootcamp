```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
df
```

```python
df.fillna(0, inplace = True)
```

```python
df['Survived'] = df['Survived'].map({
    0: 'Died',
    1: 'Survived'
})
```

```python
df['Embarked'] = df['Embarked'].map({
    'C':'Cherbourg',
    'Q':'Queenstown',
    'S':'Southampton',
})
```

```python
df.head()
```

```python
ax = sns.countplot(x = 'Pclass', hue = 'Survived', palette = 'Set1', data = df)
ax.set(title = 'Passenger status (Survived/Died) against Passenger Class', 
       xlabel = 'Passenger Class', ylabel = 'Total')
plt.show()
```

```python
print(pd.crosstab(df["Sex"],df.Survived))
ax = sns.countplot(x = 'Sex', hue = 'Survived', palette = 'Set1', data = df)
ax.set(title = 'Total Survivors According to Sex', xlabel = 'Sex', ylabel='Total')
plt.show()
```

```python
interval = (0,18,35,60,120)
```

```python
categories = ['Children','Teens','Adult', 'Old']
df['Age_cats'] = pd.cut(df.Age, interval, labels = categories)
```

```python
ax = sns.countplot(x = 'Age_cats',  data = df, hue = 'Survived', palette = 'Set1')
ax.set(xlabel='Age Category', ylabel='Total',
       title="Age Category / Survival Distribution")

plt.show()
```

```python
print(pd.crosstab(df['Embarked'], df.Survived))
ax = sns.countplot(x = 'Embarked', hue = 'Survived', palette = 'Set1', data = df)
ax.set(title = 'Survival distribution according to Port of Embarkation')
plt.show()
```
